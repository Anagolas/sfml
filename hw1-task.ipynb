```python
def limits(x):
    if x <= 10000:
        return 'A'
    elif 10000 < x <= 100000:
        return 'B'
    elif 100000 < x <= 200000:
        return 'C'
    elif 200000 < x <= 400000:
        return 'D'
    elif 400000 < x <= 700000:
        return 'E'
    elif x > 700000:
        return 'F'
```

```python
DataFrame = train['LIMIT_BAL'].map(limits)
```
